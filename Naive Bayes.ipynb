##Import các thư viện cần thiết

```python
from sklearn.naive_bayes import GaussianNB
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score
```

##Đọc và chuẩn hóa dữ liệu

```python
data = pd.read_csv("/content/breast-cancer-wisconsin.data", header = None)
# Chuyển đổi toàn bộ dữ liệu về kiểu số, các giá trị không thể chuyển sẽ thành NaN
data = data.apply(pd.to_numeric, errors='coerce')
data.fillna(data.mean(), inplace=True)
data
```

```python
df_2 = data[data.iloc[:, -1] == 2]
df_4 = data[data.iloc[:, -1] == 4]
#Chọn 80 mẫu lành tính, 40 ác tính
df_test_2 = df_2.sample(n=80, random_state= 1)
df_test_4 = df_4.sample(n=40, random_state= 1)
df_test = pd.concat([df_test_2, df_test_4])

#Phần còn lại chọn làm dữ liệu training
df_train = data.drop(df_test.index)

print(df_train)
print(df_test)
print(data.shape[0])
print(df_test.shape[0])
print(df_train.shape[0])
```

```python
# Tách đặc trưng (X) và nhãn (y)
X_train = df_train.iloc[:, :-1]  # Các cột ngoại trừ cột cuối
y_train = df_train.iloc[:, -1]   # Cột cuối là label

X_test = df_test.iloc[:, :-1]
y_test = df_test.iloc[:, -1]

print(X_train)
print(y_train)
```

```python
# Khởi tạo mô hình Gaussian Naive Bayes
gnb = GaussianNB()

# Huấn luyện mô hình với dữ liệu huấn luyện (X_train là các đặc trưng, y_train là nhãn)
gnb.fit(X_train, y_train)
```

```python
# Dự đoán trên tập kiểm tra
y_pred = gnb.predict(X_test)

# Tính toán các chỉ số
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, pos_label=2)  # `average='binary'` cho bài toán 2 lớp
recall = recall_score(y_test, y_pred, pos_label=2)

# In kết quả
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
```
